```python
# seed the pseudorandom number generator
from random import seed
seed(99)
```

```python
l = []
import random
for i in range(1000):
    l.append(random.randint(1,1000))
print(l)
```

```python
for i in set(l):
    print(i, l.count(i))
```

```python
max(l)
```

```python
min(l)
```
